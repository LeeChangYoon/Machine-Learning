## 5. Modifying Data Frames

```python
df1 = pd.read_csv('./graduate_apply.csv', sep=',', skipinitialspace=True, engine='python')
```

```python
# Select & correct by index and column name.
df1.loc[1, 'col2'] = 'Bronze and Gold' 
df1.loc[2, 'col3'] = 'í—ˆí—ˆ'
df1
```

```python
# Select & corredt by number of row and column.
df1.iloc[1, 1] = 'Bronze again' 
df1
```

```python
df1.loc[1:3, 'col2':'col3']
```

```python
# Replaces the column with the array. It could be a numpy array or a simple list.
# Could also be used to create new columns
df1.loc[1:3, 'col1'] = ['Unknown0'] * 3
df1
```

```python
df1.col3
df1.col3 = ['Unknown1'] * 4
df1.col3 
```

```python
# Equivalent to the previous
df1.col3 = ['Unknown1'] * len(df1) 
df1
```

```python
def f(x):
    return x + ' New Column';

# Uses the unary function f to create a new column based on an existing one
df1['col4'] = f(df1.col3) 
df1
```

```python
def g(x, y):
    return x + '_' + y

# Uses the 2-arg function g to create a new column based on 2 existing columns
df1['col4'] = g(df1.col3, df1.col2)
df1
```

```python
df = pd.read_csv('./graduate_apply.csv', sep=',', skipinitialspace=True, engine='python')
print(df.head())
print("------")

# Get the dummy variable from specific column.
df_rank = pd.get_dummies(df['rank'])
print(df_rank.head())
print("------")

# Concatenate the dummy data with original data.
df_new = pd.concat([df, df_rank], axis=1)
print(df_new.head())
print("------")

# Drop specific column.
df_new.drop('rank', axis=1, inplace=True)
print(df_new.head())
print("------")

# Drop multiple columns.
df_new.drop(['gre', 'gpa'], axis=1, inplace=True)
print(df_new.head())
print("------")
```
